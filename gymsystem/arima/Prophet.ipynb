```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pmdarima import auto_arima
from statsmodels.tsa.statespace.sarimax import SARIMAX

df_data = pd.read_csv("gymdata-10000.csv")
df_data['Date'] = pd.to_datetime(df_data['Date'])

# Extract the year and month from the 'Date' column
df_data['Year'] = df_data['Date'].dt.year
df_data['Month'] = df_data['Date'].dt.month

# Group and count the records for each year and month
monthly_counts = df_data.groupby(['Year', 'Month']).size().reset_index(name='Count')

# Create the x-axis values
months = np.arange(1, 13)

# Get the latest year
latest_year = df_data['Year'].max()

# Filter the data for training and test based on the latest year
train_data = monthly_counts.loc[monthly_counts['Year'] < latest_year]['Count'].values
test_data = monthly_counts.loc[monthly_counts['Year'] == latest_year]['Count'].values

# Concatenate all years' data for overall count
all_years_data = monthly_counts['Count'].values

# Use auto_arima to determine the best SARIMAX model with seasonal parameters
model = auto_arima(all_years_data, seasonal=True, m=12)

# Print the model summary
print(model.summary())

# Fit the SARIMAX model using the training data
model_fit = SARIMAX(train_data, order=model.order, seasonal_order=model.seasonal_order).fit()

# Get the forecasted values for the test period
forecast = model_fit.forecast(steps=len(test_data))

# Plot the data
bar_positions_train = np.arange(1, 13) - 0.3
bar_positions_test = np.arange(1, 13)
bar_positions_forecast = np.arange(1, 13) + 0.3

plt.figure(figsize=(8, 4))
if len(np.unique(df_data['Year'])) - 1 > 1:
    train_label = f'Train ({df_data["Year"].min()} - {latest_year - 1})'
else:
    train_label = f'Train ({latest_year - 1})'
plt.bar(bar_positions_train, train_data, width=0.3, label=train_label)
plt.bar(bar_positions_test, test_data, width=0.3, label=f'Test ({latest_year})')
plt.bar(bar_positions_forecast, forecast, width=0.3, label=f'Forecast ({latest_year})')
plt.xlabel('Month')
plt.ylabel('Count')
plt.title(f'Monthly Records Count Forecast Accuracy for Test Year ({latest_year})')
plt.legend()
plt.xticks(np.arange(1, 13), months)  # Set the x-ticks as the month values
plt.tight_layout()
plt.show()

# Use the fitted model to predict the values for future years
forecast_years = range(latest_year, latest_year + 3)

for year in forecast_years:
    forecast = model_fit.forecast(steps=12)
    forecast_all = forecast
    
    # Plot the forecasted values for the current year
    plt.figure(figsize=(8, 4))
    plt.bar(np.arange(1, len(forecast_all) + 1), forecast_all)
    plt.xlabel('Month')
    plt.ylabel('Count')
    plt.title(f'Monthly Records Count Forecast ({year})')
    plt.xticks(np.arange(1, len(forecast_all) + 1), months)  # Set the x-ticks as the month values
    plt.tight_layout()
    plt.show()

    model_fit = SARIMAX(np.concatenate([train_data, forecast]), order=model.order, seasonal_order=model.seasonal_order).fit()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Calculate MAE
mae = mean_absolute_error(test_data, forecast)
mae
```

```python
# Calculate RMSE
rmse = np.sqrt(mean_squared_error(test_data, forecast))
rmse
```

```python
# Calculate MAPE
mape = np.mean(np.abs((test_data - forecast) / test_data)) * 100
mape
```
